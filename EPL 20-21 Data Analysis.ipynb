### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
epl_df = pd.read_csv('EPL_20_21.csv')
```

```python
epl_df.head()
```

```python
epl_df.info()
```

```python
epl_df.describe()
```

### Checking Null Values

```python
epl_df.isnull().sum()
```

```python
epl_df['MinsPerMatch'] = (epl_df['Mins']/epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch'] = (epl_df['Goals']/epl_df['Matches']).astype(float)
```

```python
epl_df.head()
```

### Total Goals scored in EPL 20-21

```python
Total_Goals = epl_df['Goals'].sum()
Total_Goals
```

### Total Penalty Goals scored in EPL 20-21

```python
Total_PenaltyGoals = epl_df['Penalty_Goals'].sum()
Total_PenaltyGoals
```

###### Conclusion: Out of 986 goals scored in the EPL 2020-2021 season 102 goals were penalties

```python
Total_PenaltyAttempts = epl_df['Penalty_Attempted'].sum()
Total_PenaltyAttempts
```

###### Conclusion: Out of 125 penalties attempted 102 were converted and the rest missed

### Pie Chart for Penalties scored vs Penalties missed 

```python
plt.figure(figsize=(12,6))
pl_not_scored = epl_df['Penalty_Attempted'].sum()-Total_PenaltyGoals
data = [pl_not_scored, Total_PenaltyGoals]
labels = ['Penalties Missed', 'Penalties Scored']
color = sns.color_palette('Accent')
plt.pie(data, labels = labels, colors=color, autopct='%.0f%%')
plt.show()
```

###### Conclusion: 82% of penalties were scored and 18% were missed

### Total number of players from each country

```python
nationality = epl_df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar', figsize=(12,6), color = sns.color_palette('Paired'))
plt.xlabel("Country")
plt.ylabel("Total Number of Players")
sns.set_style("darkgrid")
plt.show()
```

### Top 5 Clubs with Most Number of Players

```python
plt.figure(figsize=(12,6))
epl_df['Club'].value_counts().head(5).plot(kind = 'bar',color = sns.color_palette('Set2'))
plt.xlabel("Club Name")
plt.ylabel("Number of Players")
plt.title("Top 5 Clubs with Most Number of Players")
sns.set_style("darkgrid")
plt.show()
```

### Top 5 Clubs with Least Number of Players

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12,6))
epl_df['Club'].value_counts().sort_values(ascending = True).head(5).plot(kind = 'bar',color = sns.color_palette('Set3'))
plt.xlabel("Club Name")
plt.ylabel("Number of Players")
plt.title("Top 5 Clubs with Least Number of Players")
plt.show()
```

### Players based on age group

```python
lessThan20 = epl_df[epl_df['Age']<20].count()['Name']
between20And25 = epl_df[(epl_df['Age']>=20) & (epl_df['Age']<25)].count()['Name']
between25And30 = epl_df[(epl_df['Age']>=25) & (epl_df['Age']<30)].count()['Name']
greaterThan30 = epl_df[epl_df['Age']>=30].count()['Name']
```

```python
data = [lessThan20, between20And25, between25And30, greaterThan30]
label = ['Age<20', 'Age>=20 and <25', 'Age>=25 and <30', 'Age>30']
plt.figure(figsize=(12,6))
plt.pie(data,labels=label,colors=sns.color_palette('Paired'),autopct='%.2f%%')
plt.title("Players based on Age")
plt.show()
```

### Total under 20 Players in Each Club

```python
sns.set_style('whitegrid')
under20 = epl_df[epl_df['Age']<20]
plt.figure(figsize=(16,6))
under20['Club'].value_counts().plot(kind = 'bar',color = sns.color_palette('cubehelix'))
plt.xlabel("Club")
plt.ylabel("Number of Players")
plt.title("Total Number of Players in each club whose age is under 20")
plt.show()
```

### Total Assist done by each club
#### Box Plot

```python
plt.figure(figsize=(20,5), dpi = 200)
sns.barplot(data = epl_df, x = 'Club', y = 'Assists', estimator=np.sum)
plt.xticks(rotation = 90)
plt.show()
```

#### Bar Plot

```python
assist = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Assists'].sum())
plt.figure(figsize=(20,5), dpi = 200)
sns.barplot(data = assist.sort_values(by = 'Assists', ascending = False), x = 'Club', y = 'Assists', palette='Set1')
plt.xticks(rotation = 90)
plt.ylabel("Total Assists")
plt.title("Total Assists done by each clubs")
plt.show()
```

### Pie Chart - Goals with Assist and without Assist

```python

plt.figure(figsize=(10,5),dpi = 150)
assists = epl_df['Assists'].sum()
data = [Total_Goals-assists, assists]
labels = ['Goals without assists', 'Goals with assists']
plt.pie(data, labels=labels, autopct='%.0f%%')
plt.title("Pie Chart for Goals with and without Assists")
plt.show()
```
